```
import ColabTurtle.Turtle as t

def polygon(length, n):
    for _ in range(n):
        t.forward(length)
        t.left(360 / n)

t.initializeTurtle()
t.penup()
t.goto(400,400)
t.speed(10)
for i in range(60):
  t.penup()
  t.forward(400-5*i)
  t.left(135.5)
  t.pendown()
  polygon(20,6)
```
